```python
import pandas as pd
```

```python
# Read in file and view
wr_te_path = ("../Resources/TE_WR.csv")
df = pd.read_csv(wr_te_path)
df
```

```python
# Check for nulls in the Position columns and Fumble columns(play is a RDE not WR or TE so stats do not matter for that player)
df.isnull().sum()
```

```python
# Remove nulls from Position columns
df.dropna(inplace=True)
```

```python
# Set WR and TE values to upper case in the Position column
df.loc[df['Position'] == 'te', 'Position'] = 'TE'
df.loc[df['Position'] == 'wr', 'Position'] = 'WR'
df
```

```python
# Fiter dataframe to WR
WR_df = df.loc[df['Position'] == "WR"]
WR_df.head(10)
```

```python
# Fiter dataframe to TE
TE_df = df.loc[df['Position'] == "TE"]
TE_df.head(10)
```

```python
# Push WR and TE dataframes to a csv file
WR_df.to_csv('../Resources/wide_receivers.csv')
TE_df.to_csv('../Resources/tight_ends.csv')
```
